How many samples have a GarageArea greater than zero? For those samples, plot SalePrice as a function of GarageArea. What does your plot suggest about predicting the SalePrice using GarageArea?

```python
SalesPrice = housing_df[housing_df.columns[18]]
GarageArea = housing_df[housing_df.columns[17]]
SalesPrice = np.array(SalesPrice)
GarageArea = np.array(GarageArea)
i = 0;
for index in GarageArea:
    if (GarageArea[index] == 0):
        GarageArea = np.delete(GarageArea, index)
        SalesPrice = np.delete(SalesPrice, index)
print(GarageArea.shape)
plt.figure(figsize=(8, 6), dpi=80)
plt.figure(1)
plt.plot(GarageArea,SalesPrice,'bo')
plt.figure(figsize=(8, 6), dpi=80)
plt.figure(2)
figure2 = plt.plot(GarageArea,SalesPrice)
for index in GarageArea:
    print(GarageArea[index])
```
